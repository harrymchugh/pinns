### 1D Heat equation

```python
# setup time-space geometry
import numpy as np

num_points = 51
x_start = 0
x_stop = 1
t_start = 0
t_stop = 1 

# thermal diffusivity
alpha = 1.22e-3

# x points
x = np.linspace(start=x_start, stop=x_stop, num=num_points)
dx = x[1] - x[0]
print(dx)

# time step for numerical stability
num_timesteps = 1000
t = np.linspace(start=t_start, stop=t_stop, num=num_timesteps)
dt = 0.5 * dx**2 / alpha
sigma = alpha * dt / dx**2

# initial condition and boundary condition
u = np.sin(np.pi * x)

# plot factor
plot_factor = int(np.floor(num_timesteps / 5))

# initial plot
import matplotlib.pyplot as plt
plt.plot(x, u)
plt.show()
```

```python
def delta_heat(u):
    sigma = alpha * dt / dx**2
    return (sigma * (u[2:] - 2*u[1:-1] + u[:-2]))
```

```python
# evolve the heat over all timesteps
for tstep in range(len(t)):
    u_next = u
    u_next[1:-1] = u[1:-1] + delta_heat(u)
    if tstep % plot_factor == 0:
        plt.plot(x, u_next)
    u = u_next

plt.show()
```
